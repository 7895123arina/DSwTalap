```python
import pandas as pd
df = pd.read_csv("file0name.csv")
msgs = df.dropna(subset=['message.text'])

msgs

msgs.count()
msgs = msgs.astype({'message.id': int, 'message.text': str})
msgs = msgs.set_index('message.id')
msgs.drop(columns=["Unnamed: 0"], inplace=True)

msgs
```

```python
import requests
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english'))
stop_words

url = "https://api.meaningcloud.com/sentiment-2.1"
t = msgs[:50]
t
# t.count()

for index, row in t.iterrows():

    a = row["message.text"]
    word_tokens = word_tokenize(a)

    filtered_sentence = [
        w for w in word_tokens if not w.lower() in stop_words]

    b = (' '.join(filtered_sentence))
    print(b)
    x = {
        'key': '',
        'txt': b,
        'lang': 'en',  # 2-letter code, like en es fr ...
    }

    t['sentiment'] = requests.post(url, data=x).text
```
